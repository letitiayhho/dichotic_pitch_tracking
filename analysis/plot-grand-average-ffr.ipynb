### Plotting dB using MNEs compute_psd

```python
def get_dB(epochs, event_ids):
    # Get the correct epochs
    cond_epochs = epochs[event_ids]

    # Compute psd for epoch and baseline TAKE BY HEMISPHERE
    P = cond_epochs.average().compute_psd(tmin = 0, tmax = 0.4, fmin = 100, fmax = 300)
    P0 = cond_epochs.average().compute_psd(tmin = -0.4, tmax = 0, fmin = 100, fmax = 300)

    # Extract the data only
    P, freqs = P.get_data(return_freqs = True)
    P0 = P0.get_data()

    # Take log
    dB = 10*np.log10(P) # Cleaner this way
#     dB = 10*np.log10(P/P0)

    # Average across channels
    dB = np.mean(dB, axis = 0)
    
    return dB, freqs

def plot_dB(ax, dB, freqs, attend, stim_freq):
#     print(str(condition))
    ax.plot(freqs, dB, label = attend)
    ax.set_xlabel('frequency')
    ax.set_ylabel('dB')
    ax.axvline(stim_freq, linestyle = '--', color = 'grey')
    ax.legend(title = 'Attend')
```

### Plotting PSD using np.fft.fft()

Not really sure I can use this since if I take the log10 of the PSD values it becomes kind of uninterpretable, but I have to take the log10 in order to run parametric tests on the values.

```python
def get_fft(FS, epochs, event_ids):
    evoked = epochs[event_ids].average()
    x = evoked.get_data().flatten()
    print(np.shape(x))
    time_step = 1 / FS
    freqs = np.fft.fftfreq(x.size, time_step)
    idx = np.argsort(freqs)
    ps = np.abs(np.fft.fft(x))**2
#     ps = 10*np.log10(ps) # take log

    return ps, freqs, idx

def plot_fft(ax, ps, freqs, idx, attend, stim_freq):
    ax.plot(freqs[idx], ps[idx], label = attend)
    ax.set_xlabel('frequency')
    ax.set_ylabel('PSD')
    ax.set_xlim(0, 300)
    ax.axvline(stim_freq, linestyle = '--', color = 'grey')
    ax.legend(title = 'Attend')
```

```python
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(10, 10), layout='constrained')
ps, freqs, idx = get_fft(FS, epochs, ['211', '212', '213'])
plot_fft(ax1, ps, freqs, idx, 'True', 130)
ps, freqs, idx = get_fft(FS, epochs, ['221', '222', '223'])
plot_fft(ax3, ps, freqs, idx, 'True', 200)
ps, freqs, idx = get_fft(FS, epochs, ['231', '232', '233'])
plot_fft(ax5, ps, freqs, idx, 'True', 280)
ps, freqs, idx = get_fft(FS, epochs, ['111', '121', '131'])
plot_fft(ax2, ps, freqs, idx, 'True', 130)
ps, freqs, idx = get_fft(FS, epochs, ['112', '122', '132'])
plot_fft(ax4, ps, freqs, idx, 'True', 200)
ps, freqs, idx = get_fft(FS, epochs, ['113', '123', '133'])
plot_fft(ax6, ps, freqs, idx, 'True', 280)
ps, freqs, idx = get_fft(FS, epochs, ['111', '112', '113'])
plot_fft(ax1, ps, freqs, idx, 'False', 130)
ps, freqs, idx = get_fft(FS, epochs, ['121', '122', '123'])
plot_fft(ax3, ps, freqs, idx, 'False', 200)
ps, freqs, idx = get_fft(FS, epochs, ['131', '132', '133'])
plot_fft(ax5, ps, freqs, idx, 'False', 280)
ps, freqs, idx = get_fft(FS, epochs, ['211', '221', '231'])
plot_fft(ax2, ps, freqs, idx, 'False', 130)
ps, freqs, idx = get_fft(FS, epochs, ['212', '222', '232'])
plot_fft(ax4, ps, freqs, idx, 'False', 200)
ps, freqs, idx = get_fft(FS, epochs, ['213', '223', '233'])
plot_fft(ax6, ps, freqs, idx, 'False', 280)
plt.show()
```

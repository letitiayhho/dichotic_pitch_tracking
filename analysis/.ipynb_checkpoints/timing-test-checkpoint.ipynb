# Timing test

```python
import numpy as np
import mne
```

```python
# Load raw data and event information
raw = mne.io.read_raw_brainvision('../data/timing_test.vhdr')
events, event_ids = mne.events_from_annotations(raw)
print(event_ids)
```

```python
# Subset events by stim
ttl_idx = (events[:,2] == event_ids['Stimulus/S 31']) # true if event code matches selection
ttl_pulses = events[ttl_idx, :] # choose just events for which ttl_idx is true

# Epoch them
epochs = mne.Epochs(raw, ttl_pulses, tmin = -.05, tmax = .25, baseline = (-.05, 0.))

# Plot stim for right aux
epochs.plot_image(picks = 'Right')

# Plot stim for left aux
epochs.plot_image(picks = 'Left')
```

```python
# Compute some info for right aux
data = epochs.get_data(picks = 'Right') # just the audio channel
data = np.squeeze(data) # remove extraneous dim
thres = np.quantile(data, .95) # select threshold for stim start
latencies = [] # sound onset time - time of TTL trigger
above_thres = (data > thres) # true where data is above threshold
for trial in range(data.shape[0]):
    idx = np.argwhere(above_thres[trial, :])[0] # first index where crosses threshold
    delta = idx/epochs.info['sfreq'] + epochs.tmin # convert to time relative to trigger
    latencies.append(delta)
    
# Get offset between TTL pulse and stim
offset = np.mean(latencies)
print(offset * 1e3) # in milliseconds

# Get the jitter between TTL pulse and stim
jitter = np.std(latencies)
print(jitter * 1e3) # in milliseconds
```
